```python
from dotenv import load_dotenv
import pandas as pd
import time
from tqdm.notebook import tqdm

load_dotenv()

from controllers.compatibilidade_controller import CompatibilidadeController

controller = CompatibilidadeController()

arquivo_anuncios = r"G:\Meu Drive\Freelas\Carlos_WildHorse\anuncios_12.06.2025.xlsx"
anuncios_df = pd.read_excel(arquivo_anuncios)
```

```python
anuncios_modificados = []

for _, row in tqdm(anuncios_df.iterrows(), total=len(anuncios_df)):
    mlb = row["MLB"]
    sku = row["SKU"]
    titulo = row["TÃ­tulo"]
    compatibilidades_removidas = None
    erro = None
    try:
        resposta = controller.remover_compatibilidades(mlb)
        compatibilidades_removidas = len(resposta)
        time.sleep(.1)
    except Exception as e:
        erro = str(e)

    anuncios_modificados.append(
        dict(mlb=mlb, sku=sku, titulo=titulo, compatibilidades_removidas=compatibilidades_removidas, erro=erro))

```

```python
anuncios_modificados_df = pd.DataFrame(anuncios_modificados)
anuncios_modificados_df.to_excel("AnÃºncios com compatibilidades removidas.xlsx")
```
